```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for the first task
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
visitors_thisweek = [120, 150, 180, 200, 230, 300, 250]
visitors_lastweek = [110, 120, 150, 250, 109, 160, 200]

# Creating the plot
plt.figure(figsize=(10, 6))
plt.plot(days, visitors_thisweek, linestyle='--', color='red', marker='o', linewidth=2, label='This Week')
plt.plot(days, visitors_lastweek, linestyle='--', color='blue', marker='o', linewidth=2, label='Last Week')

plt.title('Website Visitors Comparison')
plt.xlabel('Days of the Week')
plt.ylabel('Number of Visitors')
plt.legend()

plt.show()
```

```python
# Data for the second task
scores = [46, 55, 58, 61, 62, 65, 66, 67, 68, 70,
          71, 72, 73, 74, 76, 78, 80, 82, 84, 88, 91]

# Creating the histogram
plt.figure(figsize=(10, 6))
plt.hist(scores, bins=7, color='lightblue', alpha=0.7, edgecolor='black', linestyle='--')

plt.title('Distribution of Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')

plt.show()
```

```python
# Data for the third task
hours = [1, 2, 3, 4, 5, 6, 7, 8]
grades = [50, 55, 58, 65, 70, 75, 80, 88]
attendance = [60, 65, 70, 75, 80, 85, 90, 95]

# Creating the scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(hours, grades, c=attendance, s=attendance, cmap='plasma', edgecolors='black', alpha=0.8)

plt.title('Hours vs Grades (Colored and Sized by Attendance)')
plt.xlabel('Hours')
plt.ylabel('Grades')
plt.grid(True, linestyle='--', alpha=0.5)

plt.colorbar(scatter, label='Attendance')

plt.show()
```
